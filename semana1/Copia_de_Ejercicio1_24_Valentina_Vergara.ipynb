# Taller b√°sico pandas - numpy

```python
import pandas as pd
```

```python
# Crear un dataframe que tenga 3 registros(filas) cada una debe contener el nombre de una cancion, el artista que la canta, y la duracion en segundos
cancion = pd.Series(['Orion', 'Santa', 'Chihiro'])
artista = pd.Series(['Boza', 'Blessd', 'Billie Eilish'])
duracion = pd.Series([240,180,245])
df_music = pd.DataFrame({'Cancion': cancion, 'Artista': artista, 'Duracion(s)': duracion})
```

```python
# Imprimir la cola del dataframe
df_music.tail()
```

```python
# Acceder a la columna duracion
df_music['Duracion(s)']
```

```python
# Calcular el promedio de las duraciones
promedio = df_music['Duracion(s)'].mean()
print("El promedio de las duraciones de las canciones en segundos es: "+str(promedio))
```

```python
# Obtener los nombres de las canciones que tengan duracion mayor que el promedio
df_music[df_music['Duracion(s)'] > promedio]['Cancion']
```

```python
# Agregar una columna al dataframe que contenga True si la duracion es mayor que 3 minutos
# y false si es menor
df_music['Dura mas de 3 m?'] = df_music['Duracion(s)'] > 180
df_music.head()
```

```python
# Convertir el dataframe a numpy array
array = df_music.to_numpy()
array
```

```python
# Imprimir el shape del array
array.shape
```

```python
# Hacer un reshape para convertir el arreglo de 3x3 a un arreglo de 12 elementos
array.reshape(12)
array
```
